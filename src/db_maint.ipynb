```python
# Create the craiglist parser
my_parser = cl_parser.Parser() 
```

```python
message = ''
x = 0
for l in my_parser.listings_collection.find({"listing_removed":{'$exists':True}}):         
    try:
        doc = my_parser.listings_collection.find_one_and_update(
            {'data_id' : l['data_id']},
            {'$unset':
                {
                    'listing_removed': 1
                }
                
            },upsert=True
        )
        print('Unset listing_removed ' + l['data_id'])
        x += 1
        
    except:
        message += f'{x}|'
    # break
    
    
```

```python
x = 0
for l in my_parser.listings_collection.find({"listing_price":{'$exists':True}}):         
    try:
        listing_price_int = int(str(l['listing_price']).replace('$','').replace(',',''))    
        print("Index " + str(x) + " " +  str(listing_price_int) + " data_id " + l["data_id"])
    except:
        listing_price_int = 0
        
    
    doc = my_parser.listings_collection.find_one_and_update(
        {'data_id' : l['data_id']},
        {'$set':
            {
                'listing_price': listing_price_int                
            }
            
        },upsert=True
    )
    
    x += 1
    #if x > 100:
    #    break
```
